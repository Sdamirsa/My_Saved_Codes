```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Load the data
excel_file = r"C:\Users\LEGION\Downloads\AMIR.xlsx"
df = pd.read_excel(excel_file)

# Define column names
pre_test_col_A = "PRE V"
post_test_col_A = "POST V"
pre_test_col_B = "PRE MID"
post_test_col_B = "POST MID"

# Create separate dataframes for each group
df_A = pd.DataFrame({
    'group': 'A',
    'pre_test': df[pre_test_col_A],
    'post_test': df[post_test_col_A]
})

df_B = pd.DataFrame({
    'group': 'B',
    'pre_test': df[pre_test_col_B],
    'post_test': df[post_test_col_B]
})

# Combine the dataframes
df = pd.concat([df_A, df_B]).reset_index(drop=True)

# Function to check normality
def check_normality(data):
    _, p_value = stats.shapiro(data)
    return p_value > 0.05

# Function to check homogeneity of variances
def check_homogeneity_of_variances(group1, group2):
    _, p_value = stats.levene(group1, group2)
    return p_value > 0.05

# Function to check homogeneity of regression slopes
def check_homogeneity_of_regression_slopes(df):
    model = ols('post_test ~ C(group) + pre_test + C(group):pre_test', data=df).fit()
    return model.pvalues['C(group):pre_test'] > 0.05

# Debug Function to check homogeneity of regression slopes
def check_homogeneity_of_regression_slopes(df):
    try:
        model = ols('post_test ~ C(group) + pre_test + C(group):pre_test', data=df).fit()
        print("\nModel terms:", model.model.exog_names)
        print("Available parameters:", model.pvalues.index.tolist())
        
        # Look for the interaction term
        interaction_term = next((term for term in model.pvalues.index if ':pre_test' in term), None)
        
        if interaction_term:
            p_value = model.pvalues[interaction_term]
            print(f"P-value for interaction term '{interaction_term}': {p_value}")
            return p_value > 0.05
        else:
            print("Interaction term not found in model parameters.")
            return False
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

    
    
# Check assumptions
print("Checking ANCOVA assumptions:")

# Normality
model_residuals = ols('post_test ~ group + pre_test', data=df).fit().resid
normality_residuals = check_normality(model_residuals)
print(f"1. Normality of residuals: {'Met' if normality_residuals else 'Not met'}")
print(f"   Shapiro-Wilk test p-value: {stats.shapiro(model_residuals)[1]:.4f}")

# Homogeneity of variances
homogeneity_variances = check_homogeneity_of_variances(
    df[df['group'] == 'A']['post_test'], 
    df[df['group'] == 'B']['post_test']
)
print(f"2. Homogeneity of variances: {'Met' if homogeneity_variances else 'Not met'}")
print(f"   Levene's test p-value: {stats.levene(df[df['group'] == 'A']['post_test'], df[df['group'] == 'B']['post_test'])[1]:.4f}")

# Homogeneity of regression slopes
homogeneity_slopes = check_homogeneity_of_regression_slopes(df)
if homogeneity_slopes is not None:
    print(f"3. Homogeneity of regression slopes: {'Met' if homogeneity_slopes else 'Not met'}")
    if not homogeneity_slopes:
        print("   Note: This violation may affect the interpretation of your ANCOVA results.")
else:
    print("3. Homogeneity of regression slopes: Could not be determined due to an error")
    
print(f"3. Homogeneity of regression slopes: {'Met' if homogeneity_slopes else 'Not met'}")




# Function to add jitter to data points
def jitter(arr):
    stdev = .01 * (max(arr) - min(arr))
    return arr + np.random.randn(len(arr)) * stdev

# Linear relationship (visual check)
plt.figure(figsize=(8, 8))  # Square figure

# Determine the range for both axes
min_val = min(df['pre_test'].min(), df['post_test'].min())
max_val = max(df['pre_test'].max(), df['post_test'].max())
buffer = (max_val - min_val) * 0.1  # Add 10% buffer
plot_min = min_val - buffer
plot_max = max_val + buffer

# Create scatter plot with jittered points
for group, color in zip(['A', 'B'], ['blue', 'orange']):
    group_data = df[df['group'] == group]
    plt.scatter(jitter(group_data['pre_test']), 
                jitter(group_data['post_test']), 
                label=f'Group {group}', 
                alpha=0.7, 
                color=color)

# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# Set the same limits for both axes
plt.xlim(plot_min, plot_max)
plt.ylim(plot_min, plot_max)

# Add 45-degree line
plt.plot([plot_min, plot_max], [plot_min, plot_max], 'r--', label='x=y')

plt.xlabel('Pre-test')
plt.ylabel('Post-test')
plt.legend()
plt.title('Pre-test vs Post-test Scatter Plot \n A: V; B: MID')

# Ensure ticks are the same for both axes
ticks = np.linspace(plot_min, plot_max, 6)
plt.xticks(ticks)
plt.yticks(ticks)

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("4. Linear relationship: Please check the scatter plot for visual confirmation.")



# Perform ANCOVA
model = ols('post_test ~ C(group) + pre_test', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nANCOVA Results:")
print(anova_table)

# Post-hoc test (if needed)
tukey_results = pairwise_tukeyhsd(df['post_test'], df['group'])
print("\nTukey's HSD Test Results:")
print(tukey_results)

# Print model summary for more detailed results
print("\nDetailed Model Summary:")
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Assuming df is your dataframe from the previous analysis
# If not, load your data here

# Create a new dataframe for the differences
df_diff = pd.DataFrame({
    'group': df['group'],
    'difference': df['post_test'] - df['pre_test']
})

# Create the box plot
plt.figure(figsize=(10, 6))
box_plot = df_diff.boxplot(column='difference', by='group', notch=True)
plt.title('Difference between Post-test and Pre-test Scores by Group')
plt.suptitle('')  # This removes the automatic suptitle added by pandas
plt.ylabel('Post-test - Pre-test Score')
plt.axhline(y=0, color='r', linestyle='--')  # Add a reference line at y=0
plt.show()

# Perform paired t-tests for each group
groups = df['group'].unique()
for group in groups:
    group_data = df[df['group'] == group]
    t_stat, p_value = stats.ttest_rel(group_data['post_test'], group_data['pre_test'])
    print(f"\nPaired t-test results for group {group}:")
    print(f"t-statistic: {t_stat}")
    print(f"p-value: {p_value}")
    print(f"Mean difference: {group_data['post_test'].mean() - group_data['pre_test'].mean()}")

# Calculate and print overall statistics
print("\nOverall statistics:")
print(df_diff.groupby('group')['difference'].describe())
```
