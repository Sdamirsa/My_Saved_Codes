```python
import pygraphviz as pgv
import matplotlib.pyplot as plt

# Create a new directed graph
G = pgv.AGraph(strict=False, directed=True)

# First layer nodes
first_layer_nodes = [
    "Demographic Info", "Job", "Properties", "Medical History", "Cancer History",
    "Smoking", "Alcohol History", "Oral Health", "Physical Activity", "Tea",
    "Anthropometric", "Blood Pressure"
]

# Second layer nodes
second_layer_nodes = [f"Node {i}" for i in range(12)] + ['...', 'Node X']

# Third layer nodes
third_layer_nodes = [
    "Cancer(s) Risk", "Cardiac Event Risk", "Chronic Kidney Disease Risk",
    "Diabetes Risk", "Chronic Respiratory Diseases"
]

# Add nodes for the first, second, and third layers
fontsize = 20  # You can change it to increase font size
fontname = 'bold'

# Assign ranks in the order of the list for each layer
for i, node in enumerate(first_layer_nodes):
    G.add_node(node, shape='ellipse', style='filled', fillcolor='lightblue', fontname=fontname, fontsize=fontsize, width=2.5, height=1.2)
    G.get_node(node).attr['rank'] = str(i)

for i, node in enumerate(second_layer_nodes):
    G.add_node(node, shape='ellipse', style='filled', fillcolor='lightgrey', fontname=fontname, fontsize=fontsize, width=2.5, height=1.2)
    G.get_node(node).attr['rank'] = str(i)

for i, node in enumerate(third_layer_nodes):
    G.add_node(node, shape='ellipse', style='filled', fillcolor='#ff9248', fontname=fontname, fontsize=fontsize, width=2.5, height=1.2)
    G.get_node(node).attr['rank'] = str(i)

# Create invisible edges to specify node order within each rank
def add_invisible_edges(layer_nodes):
    for i in range(len(layer_nodes) - 1):
        G.add_edge(layer_nodes[i], layer_nodes[i + 1], style='invis')

# Connect first layer nodes to second layer nodes
for first_node in first_layer_nodes:
    for second_node in second_layer_nodes:
        G.add_edge(first_node, second_node)

# Connect second layer nodes to third layer nodes
for second_node in second_layer_nodes:
    for third_node in third_layer_nodes:
        G.add_edge(second_node, third_node)

# Set the layout to left to right alignment with centering
G.layout(prog='dot')

# Set the 'rankdir' attribute for left-to-right alignment
G.graph_attr['rankdir'] = 'LR'

# Set the 'ranksep' attribute to increase the distance between layers
G.graph_attr['ranksep'] = 6  # You can adjust this value as needed

# Save the graph to a file (optional)
graph_file = "Connected_Nodes_wPygraphviz.png"
G.draw(graph_file, prog='dot', format='png')

# Display the graph with improved centering
plt.figure(figsize=(40, 40))  # Adjust the figure size for better centering
plt.axis('off')
plt.title("Connected Nodes Graph")
plt.imshow(plt.imread(graph_file))
plt.show()
```

```python
import networkx as nx
import pygraphviz as pgv
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# First layer nodes
first_layer_nodes = [
    "Demographic", "Job", "Properties", "Medical History", "Cancer History",
    "Smoking", "Alcohol History", "Oral Health", "Physical Activity", "Tea",
    "Anthropometric", "Blood Pressure"
]

# Second layer nodes
second_layer_nodes = [f"Node {i+1}" for i in range(24)]

# Third layer nodes
third_layer_nodes = [
    "Cancer(s) Risk", "Cardiac Event Risk", "Chronic Kidney Disease Risk",
    "Diabetes Risk", "Chronic Respiratory Diseases"
]

# Add nodes for the first, second, and third layers
G.add_nodes_from(first_layer_nodes, layer=1)
G.add_nodes_from(second_layer_nodes, layer=2)
G.add_nodes_from(third_layer_nodes, layer=3)

# Connect first layer nodes to second layer nodes
for first_node in first_layer_nodes:
    for second_node in second_layer_nodes:
        G.add_edge(first_node, second_node)

# Connect second layer nodes to third layer nodes
for second_node in second_layer_nodes:
    for third_node in third_layer_nodes:
        G.add_edge(second_node, third_node)

# Define node positions for each layer
pos = {}
for i, node in enumerate(first_layer_nodes):
    pos[node] = (i * 2, 6)
for i, node in enumerate(second_layer_nodes):
    pos[node] = (i * 2, 3)
for i, node in enumerate(third_layer_nodes):
    pos[node] = (i * 2, 0)

# Create a circular layout
node_shape = 'o'  # Circular nodes

# Draw the graph
plt.figure(figsize=(24, 16))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, node_shape=node_shape)
plt.title("Connected Nodes Graph")
plt.show()
```
